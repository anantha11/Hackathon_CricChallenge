```python
matches = pd.read_csv('E:/Upgrad_DataScience/Hackathon/DataSet/Merged/MatchesInnbyInn_Master.csv')
matches.Inns = matches.Inns.apply(lambda x:int(x))
code = {'won':1,'lost':-1,'n/r':0}
matches['Result_bi'] = matches['Result'].map(code)
matches['Date'] = pd.to_datetime(matches['Date'])
matches['Year'] = matches['Date'].apply(lambda x:x.year)
matches['Overs'] = matches['Overs'].apply(lambda x:float(x))
matches['RPO'] = matches['RPO'].str.replace('-','0',case=False,regex=False)
matches['RPO'] = matches['RPO'].apply(lambda x:float(x))
matches['Score'] = matches['Score'].str.replace('TDNB','-1',case=False,regex=False)
matches['Score'] = matches['Score'].str.replace('DNB','-1',case=False,regex=False)
matches['Wickets'] = matches['Score'].str.split('/',expand=True)[1]
matches['Score'] = matches['Score'].str.split('/',expand=True)[0]
matches['Wickets'] = matches['Wickets'].fillna('10')
matches['Wickets'] = matches['Wickets'].apply(lambda x:int(x))
matches['Score'] = matches['Score'].apply(lambda x:int(x))
matches.info()
matches.head()
```

```python
seriesIndvsAus = pd.read_csv('E:/Upgrad_DataScience/Hackathon/DataSet/Merged/SeasonIndVsAus_Master.csv')
seriesIndvsAus['Season'] = seriesIndvsAus['Season'].astype(str).str.split('/',expand=True)[0]
seriesIndvsAus['Season'] = pd.to_datetime(seriesIndvsAus['Season'])
seriesIndvsAus['Year'] = seriesIndvsAus['Season'].apply(lambda x:x.year)
seriesIndvsAus['LS'] = seriesIndvsAus['LS'].str.replace('-','0',regex=False)
seriesIndvsAus['LS'] = seriesIndvsAus['LS'].apply(lambda x:int(x))
seriesIndvsAus['Win%'] = round(100*(seriesIndvsAus['Won']/seriesIndvsAus['Mat']),1)
newrow = seriesIndvsAus.loc[(seriesIndvsAus['Year'] == 2010)].copy()
newrow.Year = pd.to_numeric('2011',downcast='integer')
newrow.Season = pd.to_datetime('2011-01-01')
seriesIndvsAus = seriesIndvsAus.append(newrow,ignore_index=True)
newrow = seriesIndvsAus.loc[(seriesIndvsAus['Year'] == 1997)].copy()
newrow.Year = pd.to_numeric('1998',downcast='integer')
newrow.Season = pd.to_datetime('1998-01-01')
seriesIndvsAus = seriesIndvsAus.append(newrow,ignore_index=True)
newrow = seriesIndvsAus.loc[(seriesIndvsAus['Year'] == 2000)].copy()
newrow.Year = pd.to_numeric('2001',downcast='integer')
newrow.Season = pd.to_datetime('2001-01-01')
seriesIndvsAus = seriesIndvsAus.append(newrow,ignore_index=True)
pd.DataFrame(seriesIndvsAus[seriesIndvsAus.Team=='India'].sort_values(by='Year',ascending=False))
seriesIndvsAus.info()
seriesIndvsAus.head(10)
```

```python
#Import models from scikit learn module:
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics

#Generic function for making a classification model and accessing performance:
def classification_model(model, data, predictors, outcome):
  model.fit(data[predictors],data[outcome])
  predictions = model.predict(data[predictors])
  kf = KFold(data.shape[0], n_folds=5)
  error = []
  for train, test in kf:
    train_predictors = (data[predictors].iloc[train,:])
    train_target = data[outcome].iloc[train]
    model.fit(train_predictors, train_target)
    error.append(model.score(data[predictors].iloc[test,:], data[outcome].iloc[test]))
 
  print('Cross-Validation Score : %s' % '{0:.3%}'.format(np.mean(error)))
  model.fit(data[predictors],data[outcome])

model = RandomForestClassifier(n_estimators=100)
outcome_var = ['Result_bi']
predictor_var = ['Ground','Ind_Rating','Aus_Rating']
classification_model(model, indMatches_Master,predictor_var,outcome_var)
imp_input = pd.Series(model.feature_importances_, index=predictor_var).sort_values(ascending=False)
print(imp_input)

Team='India'
Opp='Australia'
input=['6','8.5','7']
input = np.array(input).reshape((1, -1))
outcome=model.predict(input)
print('\n\nMatch  Result')
print(list(dicVal.keys())[list(dicVal.values()).index(outcome)])
input=['2','8.5','7']
input = np.array(input).reshape((1, -1))
outcome=model.predict(input)
print('\nMatch  Result')
print(list(dicVal.keys())[list(dicVal.values()).index(outcome)])
input=['5','9.5','7']
input = np.array(input).reshape((1, -1))
outcome=model.predict(input)
print('\nMatch  Result')
print(list(dicVal.keys())[list(dicVal.values()).index(outcome)])
input=['3','9.5','7']
input = np.array(input).reshape((1, -1))
outcome=model.predict(input)
print('\nMatch  Result')
print(list(dicVal.keys())[list(dicVal.values()).index(outcome)])
input=['8','9.5','7']
input = np.array(input).reshape((1, -1))
outcome=model.predict(input)
print('\nMatch  Result')
print(list(dicVal.keys())[list(dicVal.values()).index(outcome)])
print('\n\n India - 4 Australia - 1')
```
